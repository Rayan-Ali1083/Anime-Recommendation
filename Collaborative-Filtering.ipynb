```python
import pandas as pd
import numpy as np
```

```python
anime = pd.read_csv('anime.csv')
rating = pd.read_csv('rating_complete.csv')
```

```python
anime.drop(['Name', 'Score', 'Genres', 'Japanese name',
       'Type', 'Episodes', 'Aired', 'Premiered', 'Producers', 'Licensors',
       'Studios', 'Source', 'Duration', 'Rating', 'Ranked', 'Popularity',
       'Members', 'Favorites', 'Watching', 'Completed', 'On-Hold', 'Dropped',
       'Plan to Watch', 'Score-10', 'Score-9', 'Score-8', 'Score-7', 'Score-6',
       'Score-5', 'Score-4', 'Score-3', 'Score-2', 'Score-1'], axis=1, inplace=True)
```

```python
rating.head()
```

```python
rating_anime = rating.merge(anime, left_on='anime_id', right_on='MAL_ID')
```

```python
rating_anime.drop(['MAL_ID'], axis=1, inplace=True)
```

```python
rating_anime.drop(['anime_id'], axis=1, inplace=True)
```

```python
average_ratings = rating_anime.groupby(by='user_id').mean('rating')
average_ratings.reset_index(inplace=True)
average_ratings.head()
```

```python
rating_anime.head()
```

```python
animes_name = rating_anime['English name'].unique()
```

```python
# rating_anime.drop_duplicates(subset=['user_id'], inplace=True)
```

```python
rating_anime = rating_anime.pivot(index='user_id', columns='English name', values='rating')
```

```python
rating_anime.head()
```

```python
for user_id, avg_rating in average_ratings.itertuples(index=False):
    rating_anime.loc[user_id, :] -= avg_rating
```

```python
rating_anime.fillna(0, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def similar_movies_func(df=rating_anime, user_anime="Death Note", animes_name=animes_name):
    
    main_df = pd.DataFrame(columns=['movie1', 'movie2', 'sim_score'])
    
    for anime_ in animes_name:
        if anime_ == user_anime:
            continue
        sim = cosine_similarity([df[user_anime]], [df[anime_]])
        new_row = pd.DataFrame({'movie1': [user_anime], 'movie2': [anime_], 'sim_score': [sim]})
        main_df = pd.concat([main_df, new_row], ignore_index=True)


    return main_df
```

```python
user_anime = input("Enter the movie name: ")
similar_animes = similar_movies_func(rating_anime, user_anime, animes_name)
```

```python
similar_animes.sort_values(by='sim_score', ascending=False, inplace=True)
```
